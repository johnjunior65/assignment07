## 과제의 조건으로 주어진 dataset입니다.

```python
import numpy as np
import matplotlib.pyplot as plt

num     = 1001
std     = 5 

# x  : x-coordinate data
# y1 : (noisy) y-coordinate data
# y2 : (clean) y-coordinate data 

def fun(x):
 
	# f = np.sin(x) * (1 / (1 + np.exp(-x))) 
	f = np.abs(x) * np.sin(x)

	return f

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-10,10,num)
y1      = fun(x)
y2      = y1 + nn * std

plt.plot(x, y1, 'b.')
plt.plot(x, y2, 'k.')
plt.show()
```

### 먼저 p가 3일때, 즉 degree가 3일때를 구해보겠습니다.

### 행렬 A를 만들어 주기위해 모든 원소가 1인 a를 만들어 주었습니다.

```python
a = np.linspace(1,1,num)
aa = np.array(a)
aaa = np.asmatrix(aa)

xx = np.array(x)
xxx = np.asmatrix(xx)

xx2 = np.array(x*x)
xxx2 = np.asmatrix(xx2)

xx3 = np.array(x*x*x)
xxx3 = np.asmatrix(xx3)


A = np.hstack((aaa.T,xxx.T,xxx2.T,xxx3.T))
```

### 제대로 행렬 A가 만들어졌는지 확인해보았습니다.

```python
display(A)
```

### $\theta$ matrix
$\begin{bmatrix}
\theta_0 \\
\theta_1 \\
\theta_2 \\
\theta_3 \\
\end{bmatrix}$

### $\theta$ matrix는 $(A^TA)^{-1}A^Ty_2$ 인것을 이용하여 값을 구하였습니다.

```python
yy2 = np.array(y2)
yyy2 = np.asmatrix(yy2)

theta3 = (A.T*A).I*A.T*yyy2.T

display(theta3)
```

## degree가 3인 $\hat{f}(x)$을 구하여 그래프로 나타냈습니다.

```python
new3 = theta3[0,0] + theta3[1,0] * x + theta3[2,0] * x * x + theta3[3,0] * x * x * x

plt.plot(x, y2, 'k.')
plt.plot(x, new3, 'b.')
plt.show()
```
